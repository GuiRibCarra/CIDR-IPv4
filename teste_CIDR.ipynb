```python
# CIDR em um IPV4
def converter_bin(binario):
  n = len(binario)
  decimal = 0
  i = 0
  resto = 0
  bin = int(binario)
  while n >=0:
    resto = bin % 10
    decimal = decimal + (resto*(2**i))
    n = n-1
    i = i+1
    bin = bin//10
  return decimal

a = int(input("Qual o número do 1 octeto?"))
bin(a)
b = int(input("Qual o número do 2 octeto?"))
bin(b)
c = int(input("Qual o número do 3 octeto?"))
bin(c)
d = int(input("Qual o número do 4 octeto?"))
bin(d)

masc=int(input("Qual mascara quer usar?(CIDR)"))

valores_bin = [128,64,32,16,8,4,2,1]
ope1, ope2, ope3, ope4 = 0,0,0,0
if masc < 8 :
	for i in range(1,masc):
		ope1 = ope1 + valores_bin[i]

if (masc >= 8) and (masc < 16):
	aux = masc - 8
	if aux == 0:
		ope1 = 255
	else :
		ope1 = 255
		for i in range(1,aux):
			ope2 = ope2 + valores_bin[i]

if (masc>=16) and (masc<24):
	aux = masc - 16
	if aux == 0:
		ope1, ope2 = 255,255
	else :
		ope1, ope2 = 255,255
		for i in range(1,aux):
			ope3 = ope3 + valores_bin[i]

if (masc >= 24) and (masc < 32):
	aux = masc - 24
	if aux == 0:
		ope1, ope2, ope3 = 255,255,255
	else :
		ope1, ope2, ope3 = 255,255,255
		for i in range(1,aux):
			ope4 = ope4 + valores_bin[i]

if masc == 32 :
	ope1, ope2, ope3, ope4 = 255,255,255,255

ope1_ = format(ope1,"b")
ope2_ = format(ope2,"b")
ope3_ = format(ope3,"b")
ope4_ = format(ope4,"b")

inverso_masc = []
b_dict = {'0':'1','1':'0'}

inverse_s = ''
for i in ope1_:
    if ope1_ == "0":
      inverse_s = "11111111"
      break
    inverse_s += b_dict[i]
inverso_masc.append(converter_bin(inverse_s))

inverse_s = ''
for i in ope2_:
    if ope2_ == "0":
      inverse_s = "11111111"
      break
    inverse_s += b_dict[i]
inverso_masc.append(converter_bin(inverse_s))

inverse_s = ''
for i in ope3_:
    if ope3_ == "0":
      inverse_s = "11111111"
      break
    inverse_s += b_dict[i]
inverso_masc.append(converter_bin(inverse_s))

inverse_s = ''
for i in ope4_:
    if ope4_ == "0":
      inverse_s = "11111111"
      break
    inverse_s += b_dict[i]
inverso_masc.append(converter_bin(inverse_s))

a2 = format(a,"b")
b2 = format(b,"b")
c2 = format(c,"b")
d2 = format(d,"b")


print("\nNúmero IPV4 inserido em binário:"+a2+"."+b2+"."+c2+"."+d2)
print("\nNúmero CIDR(mascara) inserido em binário:"+ope1_+"."+ope2_+"."+ope3_+"."+ope4_)

rede = [(a&ope1),(b&ope2),(c&ope3),(d&ope4)]
broadcast = [(a|inverso_masc[0]),(b|inverso_masc[1]),(c|inverso_masc[2]),(d|inverso_masc[3])]

print("\nNúmero Endereço de rede em binário:"+format(rede[0],"b")+"."+format(rede[1],"b")+"."+format(rede[2],"b")+"."+format(rede[3],"b"))
print("\nNúmero Broadcast em binário:"+format(broadcast[0],"b")+"."+format(broadcast[1],"b")+"."+format(broadcast[2],"b")+"."+format(broadcast[3],"b"))
print("\nNúmero Endereço de rede decimal:"+format(rede[0],"d")+"."+format(rede[1],"d")+"."+format(rede[2],"d")+"."+format(rede[3],"d"))
print("\nNúmero Broadcast em decimal:"+format(broadcast[0],"d")+"."+format(broadcast[1],"d")+"."+format(broadcast[2],"d")+"."+format(broadcast[3],"d"))
```
